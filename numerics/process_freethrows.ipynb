```python
%pylab inline
from simulation import *
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
import sys
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
matplotlib.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
import re
import matplotlib.gridspec as gridspec
import os
from palettable.colorbrewer.qualitative import Set1_5
from scipy.special import gammaln, polygamma
import pandas as pd
```

```python
ft_dat = pd.read_csv('game_freethrows.csv')
```

```python
results = []

for group, data in ft_dat.groupby(['player','season']):
    player = group[0]
    season = group[1]
    freethrows = list(data['outcome'])
    if len(freethrows) < 2: continue  
    
    stats =  infer_model(freethrows,["+","-"],0,0)

    info_hybrid = hybrid_info(freethrows,1)
    
    stats_1 = infer_model(freethrows,["+","-"],1,1)[0]
    
    h_models = evaluate_models([l for l in freethrows if len(l)>0],["+","-"],alpha=1,qbounds = [0,4])
    
    results += [{"player": player, "season": season, "h=1": h_models['LOO'][1]-h_models['LOO'][0],
                 "h=2": h_models['LOO'][2]-h_models['LOO'][0],
                 "h=3": h_models['LOO'][3]-h_models['LOO'][0],
                 "h=4": h_models['LOO'][4]-h_models['LOO'][0],
                 "h=1/2": info_hybrid['LOO']-h_models['LOO'][0],
                 'overall': stats[0][''][0], 'after make': stats_1['+'][0], 
                 'after miss': stats_1['-'][0],  'first': stats_1['0'][0], 
                 'n': sum(stats[1][''])
                }]
    
results = pd.DataFrame(results)
```

```python
results.sort_values(by='n',ascending=False).head(n=100)
results.to_csv("freethrow_computations.csv",index=False)
```
